HTE analysis is not your silver bullet for saving your experiment results. A common reaction to seeing a flat or statistically insignificant average impact is to look for variation in the impact with heterogeneous treatment effect (HTE) analysis. That is, we think that an imprecise aggregate impact is hiding substantial variation in precise heterogeneity estimates. 

Diving into more granular segments places more stress on the power calculation for the overall experiment (which you did, right?). The smaller a segment, the higher the expect impact needs to be to maintain statistical power. I show this point with simulated data below. Here, most of the time we are not finding a statistical significant impact for the half of the sample that has a larger impact. Note that DML has more statistical power only because I "unfairly" told it how to segment the population, while I did not for causal forest.

What are ways out of this? I discussed some standard approaches in this post (https://medium.com/@hjulian_73945/experimental-designs-ever-present-gremlin-statistical-noise-1140ab02f225). The most viable one is to introduce more features into your analysis. 

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

from econml.grf import CausalForest
```

## Create data generating process functions
We should be able to input the aggregate impact in standard deviations.

```python
def dgp(N=100, x_var=1, impact_in_sd=0.05, hte_form='linear'):
    '''
    All the impact to vary over a randomly distributed variable centered around zero
    Also introduce random noise.
    '''
    x1 = np.random.normal(0, x_var, N)
    
    e = np.random.normal(0,1,N)
    
    '''
    Treatment is randomly allocated at 50%-50%
    '''
    w = np.random.choice(2,N)
    
    '''
    HTE can take on different functions. 
    We do a simple linear and piece-wise functions here.
    '''
    if hte_form=='linear':
        hte = np.abs(x1)*impact_in_sd*0.5
    elif hte_form=='piece-wise':
        hte = impact_in_sd*(x1 <0)*0.5 + 1.5*impact_in_sd*(x1>=0)
    else:
        print('hte_form takes on the values of {linear, piece-wise}')
        
    y = x1 + w*hte + e
    return pd.DataFrame(data={'x1':x1, 'w':w, 'y':y, 'hte_gt':hte}) 
```

## How much statistical power does this design have?

```python
from statsmodels.stats.power import tt_ind_solve_power
```

```python
## Impacts in standard deviations
sd_test=0.2
df = dgp(N=500, impact_in_sd=sd_test,  hte_form='piece-wise')

```

```python
##MDE from the entire sample
mde_in_sd = tt_ind_solve_power(effect_size=None
        , nobs1=(df['w']==1).sum()
        , alpha=0.05
        , power=0.80
        , ratio =(df['w']==0).sum() / (df['w']==1).sum()
        , alternative='two-sided')
y_sd = df['y'].std()

##MDE from the subset with the largest impact
mde_in_sd_subset1 = tt_ind_solve_power(effect_size=None
        , nobs1=(df.loc[(df.x1>=0)]['w']==1).sum()
        , alpha=0.05
        , power=0.80
        , ratio =(df.loc[(df.x1>=0)]['w']==0).sum() / (df.loc[(df.x1>=0)]['w']==1).sum()
        , alternative='two-sided')
y_sd_subset1 = df.loc[df.x1>=0]['y'].std()

```

```python
print('Average impact for this dataset = {0:5.3f} ({1:5.3f} sd)'.format(sd_test, sd_test/y_sd))
print('MDE = {0:5.3f} ({1:5.3f} sd)'.format(mde_in_sd*y_sd, mde_in_sd))

print('Average impact for the subset dataset = {0:5.4f} ({1:5.4f} sd)'.format(sd_test*1.5, sd_test*1.5/y_sd_subset1))
print('MDE = {0:5.4f} ({1:5.4f} sd)'.format(mde_in_sd_subset1*y_sd_subset1, mde_in_sd_subset1))
```

## Even when we are under-powered, can we estimate precise HTE impacts?
We want to try this with three approaches:
1. Causal Forest
2. X-Learner where we just predict counterfactual outcomes
3. DML where the oracle tells us the segmentation.

```python
df = dgp(N=500, impact_in_sd=sd_test,  hte_form='piece-wise')
```

### Causal Forest

```python
grff=CausalForest(n_estimators=200)
grff.fit(X=np.array(df['x1']).reshape(-1,1), T=df['w'], y=df['y'])

df['grf_imp'] = grff.predict(np.array(df['x1']).reshape(-1,1) )
df['grf_se'] = grff.prediction_stderr(np.array(df['x1']).reshape(-1,1) )
df['grf_ci_05'],df['grf_ci_95']=grff.predict_interval(np.array(df['x1']).reshape(-1,1) )
df['stat_sig'] = ((df['grf_ci_05']>0) | (df['grf_ci_95']<0) )
```

```python
print('{0:2.1f}% of estimates are statistically significant at 95%'.format(100*df['stat_sig'].mean()))

fig,ax = plt.subplots(figsize=(9,3))

df.sort_values(by='hte_gt',inplace=True)
df.reset_index(inplace=True,drop=True)

stat_sig = df.loc[df['stat_sig']==1]
no_stat_sig= df.loc[df['stat_sig']==0]
ax.scatter(stat_sig.index, stat_sig['grf_imp'], color='grey',      alpha=0.5, label='Stat.Sig. Estimates')
ax.scatter(no_stat_sig.index, no_stat_sig['grf_imp'], color='coral', alpha=0.5,  label='Not Stat. Sig. Estimates')

ax.scatter(np.arange(len(df)), df['hte_gt'], color='blue', label='Ground Truth')
ax.set_xlabel('')
ax.set_ylabel('Impact')
ax.legend()

ax.plot([0,len(df)], [0, 0], color='black')
ax.set_title('Results from a Causal Forest')
plt.show()
```

### Simple X-Learner

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
```

```python
reg0 = RandomForestRegressor(n_estimators=500)

for w in [0,1]:
    reg_hat = reg0.fit(X=np.array(df.loc[df['w']==w]['x1']).reshape(-1,1)
                          ,y=df.loc[df['w']==w]['y'] )
    df['y_hat_{0}'.format(w)] = reg_hat.predict( np.array(df['x1']).reshape(-1,1)) 
    
df['xlearn_imp'] = df['y_hat_1'] - df['y_hat_0']
```

```python
fig,ax = plt.subplots(figsize=(9,3))

df.sort_values(by='hte_gt',inplace=True)
df.reset_index(inplace=True,drop=True)

ax.scatter(df.index, df['xlearn_imp'], color='magenta', alpha=0.25,  label='Estimates')

ax.scatter(np.arange(len(df)), df['hte_gt'], color='blue', label='Ground Truth')
ax.set_xlabel('')
ax.set_ylabel('Impact')
ax.legend()

ax.plot([0,len(df)], [0, 0], color='black')
ax.set_title('Results from a X-Learner')
plt.show()
```

### DML
We will pretent we know the segments. This gives DML an unfair advantage.

```python
import statsmodels.api as sm
```

```python
## Residualize y and w
reg_hat = RandomForestRegressor().fit(X=np.array(df['x1']).reshape(-1,1)
                      ,y=df['y'] )
df['y_hat'] = reg_hat.predict( np.array(df['x1']).reshape(-1,1)) 


reg_hat = RandomForestClassifier().fit(X=np.array(df['x1']).reshape(-1,1)
                      ,y=df['w'] )
df['w_hat'] = reg_hat.predict_proba( np.array(df['x1']).reshape(-1,1)) [:,1]

df['y_resid'] = df['y'] - df['y_hat']
df['w_resid'] = df['w'] - df['w_hat']
```

```python
df['x1_negative'] = (df['x1']<0)
df['x1_positive'] = (df['x1']>=0)
df['x1_negative_treat'] = df['x1_negative']*df['w_resid']
df['x1_positive_treat'] = df['x1_positive']*df['w_resid']
```

```python
ols_dml = sm.OLS(endog=df['y_resid']
                ,exog=sm.add_constant(df[['w_resid','x1','x1_negative_treat']])).fit()
```

```python
df['dml_imp'] = ols_dml.params['w_resid'] + ols_dml.params['x1_negative_treat']*(df['x1']<0)


var_1= np.abs( ols_dml.cov_params()['w_resid'].iloc[1]  )
var_2= np.abs( ols_dml.cov_params()['x1_negative_treat'].iloc[2]  )
cov_12=np.abs( ols_dml.cov_params()['x1_negative_treat'].iloc[1]  )
cov_se = np.sqrt( var_1 + var_2 + 2* cov_12 )

df['dml_se'] = np.sqrt(var_1)*(df['x1']>=0) + cov_se*(df['x1']<0)


df['dml_ci_05']=df['dml_imp']-1.96*df['dml_se']
df['dml_ci_95']=df['dml_imp']+1.96*df['dml_se']
```

```python
fig,ax = plt.subplots(figsize=(9,3))

df.sort_values(by='hte_gt',inplace=True)
df.reset_index(inplace=True,drop=True)

ax.scatter(df.index, df['dml_imp'], color='goldenrod', alpha=0.5,  label='Estimates')
ax.vlines(x = df.index, ymin=df['dml_ci_05'], ymax=df['dml_ci_95'], alpha=0.21, color='goldenrod', label='95% CI')
ax.scatter(np.arange(len(df)), df['hte_gt'], color='blue', label='Ground Truth')
ax.set_xlabel('')
ax.set_ylabel('Impact')
ax.legend()

ax.plot([0,len(df)], [0, 0], color='black')
ax.set_title('Results from DML')
plt.show()
```

## Run multiple simulations using Causal Forest and DML
Do not use X-Learner beccause it's just too noisy and it's not worth bootstrapping to get the confidence intervals.

```python
import numpy as np
import pandas as pd
from econml.grf import CausalForest
import statsmodels.api as sm


def calculate_mape(df, estimate_col, ground_truth_col):
    return np.mean(np.abs((df[estimate_col] - df[ground_truth_col]) / df[ground_truth_col])) * 100

# Function to generate datasets and train models
def simulate_and_analyze(num_simulations=100, sd_value=0.05):
    results = []

    for _ in range(num_simulations):
        # Generate dataset
        df = dgp(N=500, impact_in_sd=sd_value, hte_form='piece-wise')

        # Train Causal Forest model
        grf = CausalForest(n_estimators=200)
        grf.fit(X=np.array(df['x1']).reshape(-1, 1), T=df['w'], y=df['y'])
        df['grf_imp'] = grf.predict(np.array(df['x1']).reshape(-1, 1))
        df['grf_se'] = grf.prediction_stderr(np.array(df['x1']).reshape(-1, 1))
        df['grf_ci_05'], df['grf_ci_95'] = grf.predict_interval(np.array(df['x1']).reshape(-1, 1))
        df['grf_stat_sig'] = ((df['grf_ci_05'] > 0) | (df['grf_ci_95'] < 0))

        # Train DML model
        reg_hat = RandomForestRegressor().fit(X=np.array(df['x1']).reshape(-1, 1), y=df['y'])
        df['y_hat'] = reg_hat.predict(np.array(df['x1']).reshape(-1, 1))
        reg_hat = RandomForestClassifier().fit(X=np.array(df['x1']).reshape(-1, 1), y=df['w'])
        df['w_hat'] = reg_hat.predict_proba(np.array(df['x1']).reshape(-1, 1))[:, 1]
        df['y_resid'] = df['y'] - df['y_hat']
        df['w_resid'] = df['w'] - df['w_hat']
        df['x1_negative'] = (df['x1'] < 0)
        df['x1_positive'] = (df['x1'] >= 0)
        df['x1_negative_treat'] = df['x1_negative'] * df['w_resid']
        df['x1_positive_treat'] = df['x1_positive'] * df['w_resid']
        ols_dml = sm.OLS(endog=df['y_resid'], exog=sm.add_constant(df[['w_resid', 'x1', 'x1_negative_treat']])).fit()
        df['dml_imp'] = ols_dml.params['w_resid'] + ols_dml.params['x1_negative_treat'] * (df['x1'] < 0)
        var_1 = np.abs(ols_dml.cov_params()['w_resid'].iloc[1])
        var_2 = np.abs(ols_dml.cov_params()['x1_negative_treat'].iloc[2])
        cov_12 = np.abs(ols_dml.cov_params()['x1_negative_treat'].iloc[1])
        cov_se = np.sqrt(var_1 + var_2 + 2 * cov_12)
        df['dml_se'] = np.sqrt(var_1) * (df['x1'] >= 0) + cov_se * (df['x1'] < 0)
        df['dml_ci_05'] = df['dml_imp'] - 1.96 * df['dml_se']
        df['dml_ci_95'] = df['dml_imp'] + 1.96 * df['dml_se']
        df['dml_stat_sig'] = ((df['dml_ci_05'] > 0) | (df['dml_ci_95'] < 0))

        # Calculate percentages of statistically significant impacts
        grf_sig_x1_neg = df[df['x1'] < 0]['grf_stat_sig'].mean() * 100
        grf_sig_x1_pos = df[df['x1'] >= 0]['grf_stat_sig'].mean() * 100
        dml_sig_x1_neg = df[df['x1'] < 0]['dml_stat_sig'].mean() * 100
        dml_sig_x1_pos = df[df['x1'] >= 0]['dml_stat_sig'].mean() * 100

        # Calculate MAPE for both models
        grf_mape_x1_neg = calculate_mape(df[df['x1']<0],  'grf_imp', 'hte_gt')
        dml_mape_x1_neg = calculate_mape(df[df['x1']<0],  'dml_imp', 'hte_gt')
        grf_mape_x1_pos = calculate_mape(df[df['x1']>=0], 'grf_imp', 'hte_gt')
        dml_mape_x1_pos = calculate_mape(df[df['x1']>=0], 'dml_imp', 'hte_gt')
        
        results.append({
            'grf_sig_x1_neg': grf_sig_x1_neg,
            'grf_sig_x1_pos': grf_sig_x1_pos,
            'dml_sig_x1_neg': dml_sig_x1_neg,
            'dml_sig_x1_pos': dml_sig_x1_pos,

            'grf_power_x1_neg': 1-grf_sig_x1_neg,
            'grf_power_x1_pos': 1-grf_sig_x1_pos,
            'dml_power_x1_neg': 1-dml_sig_x1_neg,
            'dml_power_x1_pos': 1-dml_sig_x1_pos,
            
            'grf_mape_x1_neg': grf_mape_x1_neg,
            'dml_mape-x1_neg': dml_mape_x1_neg,
            'grf_mape_x1_pos': grf_mape_x1_pos,
            'dml_mape-x1_pos': dml_mape_x1_pos            
        })
        

    return pd.DataFrame(results)
```

```python

# Run the simulation
simulation_results = simulate_and_analyze(num_simulations=200
                                          , sd_value=sd_test)
display(simulation_results.describe().T)
```

```python
import matplotlib.pyplot as plt

def plot_statistical_power_comparison(simulation_results):
    fig, ax = plt.subplots(1, 2, figsize=(4, 3), sharey=True)

    # Plot for x1 < 0
    ax[0].bar(['CausalForest', 'DML'], 
              [simulation_results['grf_sig_x1_neg'].mean(), simulation_results['dml_sig_x1_neg'].mean()],
              color=['magenta', 'green'])
    ax[0].set_title('Sample with\n smaller impact')
    ax[0].set_ylabel('Statistical\nPower')

    # Plot for x1 >= 0
    ax[1].bar(['CausalForest', 'DML'], 
              [simulation_results['grf_sig_x1_pos'].mean(), simulation_results['dml_sig_x1_pos'].mean()],
              color=['magenta', 'green'])
    ax[1].set_title('Sample with\n larger impact')

    for p in [0,1]:
        ax[p].set_ylim(0,80)
        ax[p].grid()

    
    plt.tight_layout()
    plt.show()

# Example usage
plot_statistical_power_comparison(simulation_results)
```
