The purpose of this script is to show that increasing experiment duration may not decrease variance (and increase statistical power). We look at experiments where the outcome of customers that enter the experiment can become more noisy over time.

```python
import numpy as np
import pandas as pd

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

Create the data-generation function:

```python
def gen_data(N=1000, add_rate=0.01, T = 10):
    '''
    In each period, choose a random set of users to enter the experiment.
    For all customers in the experiment, change their outcome. This outcome
    is a cumulative function of the previous outcome.
    '''    
    data_out = pd.DataFrame(index=np.arange(N))
    data_stats = pd.DataFrame()
    '''
    Create outcome
    '''
    data_out['y'] = np.random.uniform(1,2,N)
    data_out['w'] = np.zeros(N)
    '''
    For each time period, choose a new set of users to be treated
    Also output the variance, treatment number, and standard error
    '''
    for t in range(T):
        data_out['tstar'] = np.random.uniform(0,1,N)
        data_out.loc[(data_out['w']==0) & (data_out['tstar'] < add_rate) , 'w'] = 1
        data_out['y'] =  data_out['w']*(data_out['y'] + np.random.uniform(0,0.15,N))
        n = data_out['w'].sum()
        sd = data_out['y'].std()
        se = sd/np.sqrt(n)
        data_stats = pd.concat([data_stats, pd.DataFrame(index=[t], data={'n':n, 'sd':sd, 'se':se})])
    return data_out, data_stats
```

```python
a = gen_data(T=50, add_rate=0.1)
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
ax[0].plot(a[1].index, a[1]['n'], color='orangered')
ax[0].set_xlabel('Experiment Duration')
ax[0].set_ylabel('# of Users in Experiment')

ax[1].plot(a[1].index, 1/a[1]['se'], color='royalblue', linestyle='dashed')
ax[1].set_xlabel('Experiment Duration')
ax[1].set_ylabel('Experiment Precision\n( Inverse of Standard Error )')

fig.tight_layout()
```

**sound byte**: Running your experiment longer does not always increase your precision and statistical power. In many AB experiment contexts, more users enter the experiment over time and a rule of thumb is that having more users in the experiment increases precision and statistical power. This may not be true. The cumulative outcome also becomes noisier over time. 

Using simulated data, you can get the pattern seen below. Suppose you are tracking the total number of clicks per user in the experiment. dWhile more users enter your experiment the longer it runs, those who are already in your experiment continue to click more. In the beginning of the experiment, precision increases when you increase experiment duration. However, precision starts decrease for longer duration.
