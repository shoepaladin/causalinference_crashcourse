The purpose of this script is to test DML's performance when estimating an OLS equation in an instrument variable setting. In this setting, we have an instrument $z$ that directly impacts the treatment status $w$ and does not directly impact the outcome $y$.

```python
import numpy as np
import pandas as pd

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression, LinearRegression, Lasso
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import r2_score

import statsmodels.api as sm
from statsmodels.sandbox.regression.gmm import IV2SLS
```


## Data and Estimation Functions

```python
def iv_data(N=400, X_dim=3, allow_non_linearity=False, atet=1, iv_value= np.random.uniform(0.25,0.40)):
    output_df = pd.DataFrame()
    for k in range(X_dim):
        output_df['x{0}'.format(k)] = np.random.normal(0,1, N)
    x_list = ['x{0}'.format(k) for k in range(X_dim)]
    
    ## Instrument
    output_df['z'] = np.random.normal(0,1,N)
    
    ## First Stage
    latent_w = np.dot( output_df[x_list], np.random.uniform(-1,1, X_dim) ) + np.random.normal(0,1, N)
    latent_w += iv_value*output_df['z']
    if allow_non_linearity==True:
        latent_w += np.dot( output_df[x_list].pow(2), np.random.uniform(-1,1, X_dim) )
        latent_w += iv_value*output_df['z']
    output_df['w'] = (latent_w > np.percentile(latent_w, 50)).astype(float)
    
    ## Second Stage, the outcome
    output_df['y'] = np.dot( output_df[x_list], np.random.uniform(-1,1, X_dim) ) + np.random.normal(0,1, N)
    output_df['y'] += atet*output_df['w']
    if allow_non_linearity==True:
        output_df['y'] += np.dot( output_df[x_list].pow(2), np.random.uniform(-1,1, X_dim) )
        
    return output_df
```

```python
def ols_naive(data=None, x_list=['x0','x1','x2']):    
    ols_naive = sm.OLS(data['y'],sm.add_constant(data[['w']+x_list])).fit()
    return [ ols_naive.params[-1], ols_naive.bse[-1]]
```

```python
## Code up a DML using just LASSO Regression for the treatment residualization. 
## Record whether the treatment is selected for the treatment and outcome residualization
def dml_sel(data=None, x_list=['x0','x1','x2']
            , t_model=LogisticRegression(fit_intercept=True, C=0.5, penalty='l1', solver='saga')
            , y_model=Lasso(fit_intercept=True, alpha=0.5) ):
    ## Do prediction and feature selection for treatment and control
    
    ## Estimate models and see whether they select different variables.
    lgf = t_model.fit(X=data[x_list+['z']], y=data['w'])
    treatment_z=[] 
    try:
        for e in zip(x_list+['z'], lgf.coef_[0]):
            if e[1]!=0:
                treatment_z.append(1)
            else:
                treatment_z.append(0)
    except:
        for e in zip(x_list+['z'], lgf.feature_importances_):
            if (e[1] > 0.1):
                treatment_z.append(1)
            else:
                treatment_z.append(0)
    
    lrf = y_model.fit(X=data[x_list+['z']], y=data['w'])
    treatment_y=[]
    try:
        for e in zip(x_list+['z'], lrf.coef_):
            if (e[1]!=0):
                treatment_y.append(1)
            else:
                treatment_y.append(0)
    except:
        for e in zip(x_list+['z'], lrf.feature_importances_):
            if (e[1] > 0.1):
                treatment_y.append(1)
            else:
                treatment_y.append(0)    
    print(lgf.coef_)
    ## Estimate DML
    w_residual = data['w'] - lgf.predict_proba(data[x_list+['z']])[:,1]
    y_residual = data['y'] - lrf.predict(data[x_list+['z']])    
    
    dml_ols = sm.OLS(y_residual,sm.add_constant(w_residual)).fit()
    return [ dml_ols.params[-1], dml_ols.bse[-1], treatment_z, treatment_y,]
```

```python
def dml_iv(data=None, x_list=['x0','x1','x2']
            , t_model=LogisticRegression(fit_intercept=True, C=0.5, penalty='l1', solver='saga')
            , y_model=Lasso(fit_intercept=True, alpha=0.5) ):
    '''
    DML using IV
    Three different residuals:
        1. Y on X
        2. D on X
        3. Z on X
    The second stage is an IV regression there the outcome is the residual from 1. , 
        and instrument is residual from 3., 
        and endogenous variable is residual from 2.
    '''
    ## Estimate models
    y_model_fit = y_model.fit(X=data[x_list], y=data['y'])
    t_model_fit = t_model.fit(X=data[x_list], y=data['w'])
    z_model_fit = y_model.fit(X=data[x_list], y=data['z'])
    
    ## Residuals
    y_residual = data['y'] - y_model_fit.predict(data[x_list])    
    t_residual = data['w'] - t_model_fit.predict_proba(data[x_list])[:,1]
    z_residual = data['z'] - z_model_fit.predict(data[x_list])
    
    
    z3=IV2SLS(exog=sm.add_constant(t_residual)
      , endog=y_residual
      , instrument=sm.add_constant(z_residual)).fit()
              
    return z3
```

## Simulations
Record whether Z is selected in each one, and correlate it with bias.

```python
simuls = 50
atet_list=[0.10,0.5, 10]
atet_list=[0.10,10]
```

```python
df_sim = pd.DataFrame()
for iv in [0.001, 0.10, 0.50]:
    for atet_ in atet_list:
        for s in range(simuls):
            df = iv_data(atet=atet_, iv_value=iv, allow_non_linearity=True)

            ## Results using DML-Lasso 
            entry = dml_sel(data=df, x_list=['x0','x1','x2'])
            sim_entry = pd.DataFrame(data=[[entry[0], entry[1]] +entry[2]+entry[3]], columns=['coef','se'
                                      ,'select_w_x0'
                                      ,'select_w_x1'
                                      ,'select_w_x2'
                                      , 'select_w_z'
                                      ,'select_y_x0'
                                      ,'select_y_x1'
                                      ,'select_y_x2'
                                      , 'select_y_z'] )
            sim_entry['gt'] = atet_

            z3=IV2SLS(exog=sm.add_constant(df[['w','x0','x1','x2']])
              , endog=df['y']
              , instrument=sm.add_constant(df[['z','x0','x1','x2']])).fit()
            sim_entry['2sls'] = z3.params[1]
            sim_entry['se_2sls'] = z3.bse[1]
            sim_entry['iv_value'] = iv 
            ## Results using OLS, ignoring the instrument problem
            ols_ = ols_naive(data=df, x_list=['x0','x1','x2','z'])
            sim_entry['coef_ols'] = ols_[0]
            sim_entry['se_ols'] = ols_[1]

            ## DML IV
            djk3=dml_iv(data=df, x_list=['x0','x1','x2'])
            sim_entry['coef_dml_iv'] = djk3.params['w']
            sim_entry['se_dml_iv'] = djk3.bse['w']
            df_sim = pd.concat([df_sim, sim_entry])
```

```python
for r in ['2sls','coef','coef_ols', 'coef_dml_iv']:
    df_sim['bias_{0}'.format(r)] =   df_sim[r] - df_sim['gt']
    df_sim['mape_{0}'.format(r)] = np.abs( (df_sim[r] - df_sim['gt'])/df_sim['gt'] )
```

```python
def remove_max_min(ar=None):
    return ar[ (ar!=np.min(ar) ) &  (ar!=np.max(ar) ) ]
def winsorize(ar=None, q=2):
    return ar[ (ar>=np.percentile(ar, q) ) &  (ar<=np.percentile(ar, 100-q) ) ]
```

```python
fig,ax = plt.subplots(ncols=len(atet_list), nrows=1,figsize=(8,2))
plot = 'mape'
plot = 'bias'
for atet_ in enumerate(atet_list):
    data_ = df_sim.loc[df_sim['gt']==atet_[1]]
    for i in zip(['2sls','coef','coef_ols','coef_dml_iv'], ['2SLS','DML','OLS','DML-IV']):
        xj3=ax[atet_[0]].hist(winsorize(data_['{0}_{1}'.format(plot,i[0])]),    alpha=0.45,    label=i[1])
    ax[atet_[0]].set_xlabel('{0} Bias'.format(plot))
    ax[atet_[0]].legend(bbox_to_anchor=(1.1, 1.05))
    ax[atet_[0]].grid()

    ax[atet_[0]].set_title('True Impact is {0}'.format(atet_[1]))
plt.show()
plt.tight_layout()

fig,ax = plt.subplots(ncols=len(atet_list), nrows=1,figsize=(8,2))
plot = 'mape'
plot = 'bias'
for atet_ in enumerate(atet_list):
    data_ = df_sim.loc[df_sim['gt']==atet_[1]]
    for i in zip(['2sls','coef','coef_ols','coef_dml_iv'], ['2SLS','DML','OLS','DML-IV']):
        data_.sort_values(by='{0}_{1}'.format(plot, i[0]),inplace=True)
        dd3=ax[atet_[0]].scatter(np.arange(len(winsorize(data_['{0}_{1}'.format(plot,i[0])])))
                                 , winsorize(data_['{0}_{1}'.format(plot,i[0])]),  s=5,   alpha=0.45,  label=i[1])
        
    ax[atet_[0]].set_ylabel('{0} Bias'.format(plot))
    ax[atet_[0]].set_xlabel('Iterations')
    ax[atet_[0]].legend(bbox_to_anchor=(1.1, 1.05))
    ax[atet_[0]].grid()



    ax[atet_[0]].set_title('True Impact is {0}'.format(atet_[1]))
plt.show()
plt.tight_layout()

```

```python
fig,ax = plt.subplots(ncols=len(atet_list), nrows=1,figsize=(8,2))

for atet_ in enumerate(atet_list):
    data_ = df_sim.loc[df_sim['gt']==atet_[1]]
    for i in zip(['se_2sls','se','se_ols','se_dml_iv'], ['2SLS','DML','OLS','DML-IV']):
        xj3=ax[atet_[0]].hist(winsorize(data_['{0}'.format(i[0])]),    alpha=0.55,    label=i[1])
    ax[atet_[0]].set_xlabel
    ax[atet_[0]].set_ylabel('{0} Standard Error'.format(i[0]))
    ax[atet_[0]].legend(bbox_to_anchor=(1.1, 1.05))
    ax[atet_[0]].grid()

    ax[atet_[0]].set_title('True Impact is {0}'.format(atet_[1]))
plt.show()
plt.tight_layout()

fig,ax = plt.subplots(ncols=len(atet_list), nrows=1,figsize=(8,2))
plot = 'mape'
plot = 'bias'
for atet_ in enumerate(atet_list):
    data_ = df_sim.loc[df_sim['gt']==atet_[1]]
    for i in zip(['se_2sls','se','se_ols','se_dml_iv'], ['2SLS','DML','OLS','DML-IV']):
        data_.sort_values(by='{0}'.format(i[0]),inplace=True)
        dd3=ax[atet_[0]].scatter(np.arange(len(winsorize(data_['{0}'.format(i[0])])))
                                 , winsorize(data_['{0}'.format(i[0])]),  s=5,   alpha=0.55,  label=i[1])
        
    ax[atet_[0]].set_ylabel('Standard Error'.format(i[0]))
    ax[atet_[0]].set_xlabel('Iterations')
    ax[atet_[0]].legend(bbox_to_anchor=(1.1, 1.05))
    ax[atet_[0]].grid()



    ax[atet_[0]].set_title('True Impact is {0}'.format(atet_[1]))
plt.show()
plt.tight_layout()

```
